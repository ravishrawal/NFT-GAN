# FFHQ

```python
networks = []
for root, dirs, files in os.walk(".", topdown=False):
    for name in files:
        prefix='network-snapshot-'
        if prefix in name:
            network_path = os.path.join(root, name)
            networks.append(network_path)
networks
```

```python
metrics = ['fid50k_full', 'kid50k_full']
for network in networks:
    for metric in metrics:
        cmd = f"""python calc_metrics.py --metrics={metric} --network={network}"""
        !$cmd
```
